### Pregunta 1.14

```python
# Las variables que más correlacionaron fueron Population, Area y Phones, sin considerar GPD per cápita
outliers = ["Population", "Area (sq. mi.)", "Phones (per 1000)"]

for i in outliers:
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3 - Q1
  low = Q1 - 1.5 * IQR #se considera outlier los que están fuera de low y high
  high = Q3 + 1.5 * IQR
  out=df[(df[i] < low) | (df[i] > high)] #valores fuera de los límites low y high
  print(f"\nOutliers en {i}:")
  print(out[['Country', i, 'GDP ($)']].to_string(index=False))
  porcentaje=(len(out)/len(df))*100 #porcentaje de los datos que son outliers
  print(f"\nPorcentaje de outliers en {i}: {porcentaje}%")
```

### Pregunta 1.15

En los ejemplos anteriores calculamos correlaciones para `GDP ($ per capita)` y `GDP ($)`. Genere un nuevo dataframe que tenga le variación porcentual de la correlación absoluta para cada una de las columnas de características, e.g., si la correlación en valor absoluto de `GDP ($ per capita)` vs `Industry` es 0.1 y la correlación `GDP ($)` vs `Industry` es 0.5, la variación deberá ser +500%. Dicha variación porcentual puede ser positiva o negativa, pero ordene los el dataframe de tal manera que la variación de correlación absoluta sea desendiente.

```python
Correlacion_PC=df[df.dtypes[df.dtypes != "object"].index].corr()["GDP ($ per capita)"]
Correlacion_PC

Correlacion_Total=df[df.dtypes[df.dtypes != "object"].index].corr()["GDP ($)"]
Correlacion_Total

df_2=Correlacion_PC.reset_index()
df_3=Correlacion_Total.reset_index()

df_nuevo=pd.merge(df_2,df_3,on='index')
df_nuevo

df_nuevo['Variacion (%)']=df_nuevo['GDP ($)']/df_nuevo['GDP ($ per capita)']*100
df_nuevo

df_nuevo['Variacion (% abs)']=abs(df_nuevo['GDP ($)']/df_nuevo['GDP ($ per capita)']*100)
df_nuevo

df_nuevo=df_nuevo.sort_values(by='Variacion (% abs)',ascending=False)

df_nuevo
```

Para cada serie de total de emisiones por país, calcule el cambio porcentual a través del tiempo. Realice imputación de missings si considera necesario, justificando su imputación. Si no lo considera necesario, también justifique (se evaluará un buen criterio fundamentado).

Repita el ejercicio del gráfico de series de tiempo anterior, pero graficando los **cambios porcentuales** para años mayores o iguales a 1995. ¿Cómo interpretaría económicamente el shock sobre las emisiones de CO2 tanto en la crisis subprime como en la crisis del Covid-19?


```python
  df_co2_año=df_co2[df_co2["Year"]>=1995]
  df_co2_año=df_co2_año.sort_values(by=["Country", "Year"])
  df_co2_año["cambio_porcentual"]=df_co2_año.groupby("Country")["Total"].pct_change()*100

  sns.lineplot(data=df_co2_año, x='Year', y='cambio_porcentual', hue='Country', legend=False)
  plt.axvline(2008, color='red', linestyle='--', label='Crisis Subprime')
  plt.axvline(2020, color='purple', linestyle='--', label='COVID-19')
  plt.title('Cambio porcentual en emisiones de CO2 por país desde 1995 (todos los países)')
  plt.ylabel('Cambio porcentual (%)')
  plt.xlabel('Año')
  plt.legend()
  plt.show()

  paises_relevantes=['United Kingdom', 'Canada', 'Germany', 'France', 'United States', 'Brazil', 'China', 'Japan', 'India']
  sns.lineplot(data=df_co2_año[df_co2_año['Country'].isin(paises_relevantes)], x='Year', y='cambio_porcentual', hue="Country")
  plt.axvline(2008, color='red', linestyle='--', label='Crisis Subprime')
  plt.axvline(2020,color="purple", linestyle="--",label="COVID-19")
  plt.title("Cambio porcentual en emisiones de CO2 por país desde 1995 (países relevantes)")
  plt.ylabel("Cambio porcentual (%)")
  plt.xlabel("Año");plt.legend()
  plt.show
```


En ambos gráficos se aprecia una caída significativa en el cambio porcentual de las emisiones luego de la Crísis Sub Prime. Esto se debe en gran parte a la recesión. Con esta última, algunos sectores redujeron su actividad y, dentro de estos, se pueden haber encontrado los sectores industriales y de transporte. Al haber menos consumo en épocas de crísis, hay menos producción por parte de las empresas como reacción ante esa menor demanda y, por ende, se usa menos energía a la hora de producir y así hay menores emisiones.

Respecto al COVID, la pandemia frenó la economía durante un tiempo. Se cancelaron vuelos, industrias cerraron, menos movilidad interurbana, etc. Todo eso produjo una menor liberación de gases de efecto invernadero.

En conclusión, ambas crísis muestran que las emisiones están relacionadas a la actividad económica mundial.
